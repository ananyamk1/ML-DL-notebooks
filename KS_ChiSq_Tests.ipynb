Kolmogorov-Smirnov Test 

```python
import numpy as np
from scipy.stats import kstest
import random
import scipy.stats as stats
import statsmodels.api as sm
```

```python
N = int(input("Enter number of random numbers: "))
actual =[]
print("Enter outcomes: ")

for i in range(N):
    actual.append(random.random())

x = kstest(actual, "norm")  
print(x)

y = kstest(actual, "uniform")
print(y)
```

```python
from numpy.random import poisson
a = poisson(5,15)
x = kstest(a, "norm")
print(x)
```

```python
#normal random variables
#rvs function returns sample from same distribution

x_20 = stats.norm.rvs(size=20)
bin = np.random.binomial(10, 0.3, size=20)
for i in [x_20]:
    print (stats.kstest(i, 'norm'))
    print (stats.kstest(i, 'uniform'))
for j in [bin]:
    print (stats.kstest(j, 'uniform'))
```

Chi Square Test

```python
from scipy.stats import chisquare
do = [30, 60, 10]
de = [50, 45, 5]
m = chisquare(de, do, ddof=2)
print(m)
```

```python
from scipy.stats import chi2_contingency
data = ([20, 30, 50], [35, 40, 25])
stat, p, dof, expected = chi2_contingency(data)
print(chi2_contingency(data))
#significance_level 
sl = 0.05
if p <= sl:
  print("Accept hypothesis")
else:
  print("Reject hypothesis") 
```
